```python
import numpy as np
import control as ctl
import matplotlib.pyplot as plt
```

### Cruise Model

```python
m = 1000
b = 50

A = -b/m
B = 1/m
C = 1
D = 0

G = ctl.ss(A, B, C, D)

input = 500
t, y = ctl.step_response(500*G)
plt.plot(t, y)
plt.xlabel('Time')
plt.title('Reponse of a Cruise System')
```

### Motor Model

```python
J = 0.01
b = 0.1
K = 0.01
R = 1
L = 0.5
s = ctl.tf([1, 0],[1])
G = K/((J*s+b)*(L*s+R)+K**2)

t, y = ctl.step_response(G)
plt.plot(t, y)
plt.xlabel('Time')
plt.title('Response of a Motor')
```
